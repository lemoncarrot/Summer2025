# SpatialData Cheat Sheet

SpatialData is a container for various SpatialElements

Images: H&E
Labels: pixel-level segmentation
Points: Locations of transcripts (matches up with Tables) or landmark points
Shapes: cell/nucleus boundaries (polygons)
Tables: matrices annotating SpatialElements

Images and Labels are Rasters
Points and Shapes are Vectors

```python
import spatialdata as sd

# Function to read zarr file
sdata = sd.read_zarr("mouse_liver.zarr")
sdata
```
